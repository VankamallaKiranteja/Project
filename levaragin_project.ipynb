```python
import pandas as pd
import random
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns

# 1. We load the dataset from the CSV file using pd.read_csv().

# 2. We create a pairplot using sns.pairplot() to visualize the relationships between different variables, including temperature, CO2 concentration, salinity, alkalinity, and pH.

# 3. We create a histogram of pH values using sns.histplot() to show the distribution of pH values in the dataset.

# 4. We create a boxplot of pH values by location using sns.boxplot() to see how pH varies across different locations in Andhra Pradesh.
```

```python
df = pd.read_csv('seawater_ph_dataset.csv')
# print("Dataset 'seawater_ph_dataset.csv' created successfully with 500 rows.")
print(df)
```

```python
# code for visualization of the date 

# Create a pairplot to visualize relationships between variables
# sns.pairplot(df, vars=['Temperature (Â°C)', 'CO2 Concentration (ppm)', 'Salinity (ppt)', 'Alkalinity (Âµmol/kg)', 'pH'])
sns.pairplot(df, vars=['Temperature (Â°C)', 'Dissolved Oxygen (ppm)', 'pH'])

plt.suptitle('Pairplot of Seawater Variables')
plt.show()
```

```python
# Create a histogram of pH values
plt.figure(figsize=(6, 6))
sns.histplot(data=df, x='pH', bins=20, kde=True)
plt.title('Distribution of pH Values')
plt.xlabel('pH')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a boxplot of pH values by location
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Location', y='pH')
plt.title('pH Values by Location')
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('pH')
plt.show()
```
